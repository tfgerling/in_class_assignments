```python
df['AmFoot']=df['American football: (United States)'].pct_change()
df['AmFoot'].plot()
plt.ylabel('Football')
plt.xlabel('Year')
plt.title('Interest Over Time - American')
plt.show()
```

```python
#Create the acf plot for the American Football column
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(df[['American football: (United States)']], lags = 10, missing='drop',alpha=0.01)
plt.title('ACF plot American Football with 99% confidence level')
plt.show()
```

```python
#Create the pacf plot for the American Football column
plot_pacf(df[['American football: (United States)']], lags = 10,alpha=0.01)
plt.title('PACF plot Amrican Football with 99% confidence level')
plt.show()
```

```python
#Create the acf plot for the Football column
plot_acf(df[['Football: (United States)']],lags = 10, missing='drop', alpha=0.01)
plt.title('ACF plot Football with 99% confidence level')
plt.show()
```

```python
#Create the pacf plot for the Football column
plot_pacf(df[['Football: (United States)']], zero = False,lags = 10,alpha=0.01)
plt.title('PACF plot Football with 99% confidence level')
plt.show()
```

## Decide whether ARMA, ARIMA or SARIMA makes more sense for the dataset

SARIMA makes more sense for the Football: (United States) data as the AIC value is lower. This makes sense since football isn't played year round so the interest would tend toward seasonal. Generally speaking, a smaller AIC means the model fits the time series better.

The SARIMA model actually had the lowest AIC, leading to the conclusion that the SARIMA made the time series stationary and is most sutiable for forecasting. 

## Run ARMA, ARIMA or SARIMA on it
- chose to run all 3

ARMA on Football: (United States) data

```python
from statsmodels.tsa.arima.model import ARIMA

#using ARIMA for ARMA since this is an updated version of statsmodels
# (1,0,5) is used because there are 5 points outside the confidence interval for MA and 1 significant one for AR

model = ARIMA(df['Football: (United States)'],order=(1,0,5))

# fit the model to the data
results = model.fit()

#print the summary results
results.summary()
```

ARIMA on Football: (United States) data

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(df['Football: (United States)'],order=(1,1,5))

# fit the model to the data
results = model.fit()

#print the summary results
results.summary()
```

SARIMA on Football: (United States) data

```python
import statsmodels.api as sm
model = sm.tsa.statespace.SARIMAX(df['Football: (United States)'],
                                  order=(1,1,5),
                                 seasonal_order=(1,1,1,12),
                                 enforce_stationarity=False,
                                 enforce_invertibility=False)

# fit the model to the data
results=model.fit()

#print the summary results
print(results.summary())
```

## Explain the results


There is 1 significant point outside the 99% confidence interval of the PACF plot (AR) and 5 points outside of the 99% confidence interval for ACF plot (MA). These values were then used to run ARMA, ARIMA and SARIMA to determine which had the lowest AIC value. SARIMA had the lowest AIC. I then looked into how to double check my thinking and came across auto_arima, which shows below. Running that confirms that the model best used is SARIMA.

```python
from pmdarima.arima import auto_arima

model = auto_arima(df['Football: (United States)'], start_p=0, start_q=0)
model.summary()
```

# What is a fourier transform? 

The Fourier Transform is a tool that breaks a waveform (a function or signal) into an alternate representation, characterized by sine and cosines. The Fourier Transform shows that any waveform can be re-written as the sum of sinusoidal functions.

Perform one on y = np.cos(x/2), and both show (with code) and explain the key elements

```python
from scipy import signal

x = np.arange(1,100,1) # to generate sequences of numbers
y = np.cos(x/2)

plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Intensity')
plt.title('Signal of y=np.cos(x/2)')
plt.show()
```

```python
#power spectral density - The power spectral density (PSD) of the signal describes the power present 
# in the signal as a function of frequency, per unit frequency.

psd = np.abs(np.fft.rfft(y))
freqs = np.fft.rfftfreq(len(y))
plt.plot(freqs,psd)

#amplitude - a measure of how much energy per unit of bandwidth you have.
plt.axvline(1/(6*np.pi),c='y') # all I understand on this line is the c means color of the amplitude line

plt.xlabel('Frequency')
plt.ylabel('Intensity')
plt.title('Fourier Transform of y=np.cos(x/2)')
plt.show()
```
