## Describe the difference between linear and rank correlation

Linear correlation measures the proximity of the mathematical relationship between variables or dataset features to a linear function. Rank correlation compares the ranks or the orderings of the data related to two variables or dataset features. Rank correlation is concerned only with the order of values, not with the particular values from the dataset.

Create a scatter plot showing the relationship between
calculated_host_listings_count and availability_365 in the Belize Airbnb data

```python
import pandas as pd
belize_df = pd.read_csv("belize_airbnb_listings.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plot1 = sns.scatterplot(data=belize_df, x="availability_365", y="calculated_host_listings_count")
plot1.set_title('Host Listings with Availability')
plot1.set(xlabel="Availability", ylabel="Host Listings")
plt.show()
```

Calculate the correlation coefficient for the data in the scatter plot, and describe
what is happening.

```python
belize_df['availability_365'].corr(belize_df['calculated_host_listings_count'])
```

There does not appear to be a strong correlation between the 2 columns
